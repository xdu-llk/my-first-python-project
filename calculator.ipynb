```python
# 简单计算器
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "不能除以0"
    return a / b

print("欢迎使用简易计算器")
print("1. 加法  2. 减法  3. 乘法  4. 除法")

choice = input("请输入选择(1/2/3/4)：")

num1 = float(input("第一个数："))
num2 = float(input("第二个数："))

if choice == "1":
    print("结果：", add(num1, num2))
elif choice == "2":
    print("结果：", sub(num1, num2))
elif choice == "3":
    print("结果：", mul(num1, num2))
elif choice == "4":
    print("结果：", div(num1, num2))
else:
    print("输入错误")
```
